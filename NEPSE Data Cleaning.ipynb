```python
# Importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# After all the Nepse Data is downloaded by web scraping on Google Collab (Damn! It took me almost 4 hours.) now, we are ready for cleaning the data. 
unclean_stock = pd.read_csv('../content/drive/My Drive/NepseData/Uncleaned_stock_data_from_2010_April_13_till_2019_June_06.csv')
```

```python
unclean_stock.shape
```

```python
unclean_stock.iloc[:2,:]
```

```python
# Taking only the data from April-19-2010. I think on April-15-2010, NEPSE was testing their web app.
clean1 = pd.DataFrame({
    #'Date' : unclean_stock.iloc[179:,1],
    'Traded Company': unclean_stock.iloc[179:,3],
    'No. of Transaction': unclean_stock.iloc[179:,4],
    'Maximum Price': unclean_stock.iloc[179:,5],
    'Minimum Price': unclean_stock.iloc[179:,6],
    'Closing Price': unclean_stock.iloc[179:,7],
    'Traded Shares': unclean_stock.iloc[179:,8],
    'Total Amount During Transaction': unclean_stock.iloc[179:,9],
    'Previous Closing Price': unclean_stock.iloc[179:,10],
    'Difference': unclean_stock.iloc[179:,11]
}).set_index(unclean_stock.iloc[179:,1]).rename_axis('Dates')


#.reset_index(drop=True)

print("You have dropped 179 data of April-15-2010")
```

```python
clean1.head()
```

```python
clean1.shape
```

```python
# Replacing the 0.0 values from the data frame with NaN then, remove the whole rows that includes NaN

clean1["Closing Price"].replace(0.0, np.NaN, inplace = True)
clean1.dropna(inplace = True)

clear_output()
```

```python
# Saving Nepse all company data from 2010 to 2019.
clean1.to_csv("BetterNepseAllCompanyData2010To2019.csv")

print("Data Saved! Nepse all company data from 2010 to 2019.")
```

```python
# Cleaning all the companies data
clean2 = pd.DataFrame({
    'Stock Name': allCompanies.iloc[:,1],
    'Stock Symbol': allCompanies.iloc[:,2],
    'Sector': allCompanies.iloc[:,3],
    'Website': allCompanies.iloc[:,4],
    'URL ID': allCompanies.iloc[:,5],
}).reset_index(drop=True)
```

```python
# Saving all company info into a csv file.
clean2.to_csv("AllCompanyInfo.csv")
```
